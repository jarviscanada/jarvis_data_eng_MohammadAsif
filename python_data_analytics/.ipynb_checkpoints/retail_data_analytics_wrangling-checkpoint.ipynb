## Imports

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sqlalchemy
```

## Data Preparation

```python
engine_string = 'postgresql://postgres:password@localhost:5432/postgres'  
engine = sqlalchemy.create_engine(engine_string)

retail_df = pd.read_sql_table('retail', con=engine)
print(retail_df.head())
```

```python
retail_df.info()
retail_df.describe()
```

```python
dtype_map = {'invoice_no': 'string',
             'stock_code': 'string',
             'description': 'string',
             'quantity': 'int64',
             'invoice_date': 'datetime64[ns]',
             'unit_price': 'float64',
             'customer_id': 'float64',
             'country': 'category'} ## as the csv has no metadata dtypes are mapped to dataframe

retail_df = pd.read_csv('retail.csv') \
.astype(dtype_map)
retail_df['total_amount'] = retail_df['quantity'] * retail_df['unit_price'] ## revenue calculation 
retail_df.head()
```

## Total Invoice Amount Distribution

```python
## Removing values that are canceled and 0 as they are not significant in result
filtered_df = retail_df[retail_df['invoice_no'].str.isdigit() & (retail_df['total_amount'] > 0)]
invoice_total_df = filtered_df.groupby('invoice_no')['total_amount'].sum()
```

```python
def graph(series):
    print('Minimum:', np.round(series.min(), 2))
    print('Mean:', np.round(series.mean(), 2))
    print('Median:', np.round(series.median(), 2))
    print('Mode:', np.round(series.mode().max(), 2))
    print('Maximum:', np.round(series.max(), 2))
    
    fig, ax = plt.subplots(2, 1)
    fig.set_size_inches(10,5)
    ax[0].set_title('Data Distribution')
    ax[0].hist(series)
    ax[0].set_ylabel('Frequency')
    ax[1].boxplot(series.dropna(), vert=False)
    plt.show()
graph(invoice_total_df)
```

```python
quant_val = invoice_total_df.quantile(0.85)
invoice_total_outliers_df = invoice_total_df.where(lambda x : x <= quant_val)
graph(invoice_total_outliers_df)
```

## Monthly Placed and Canceled Orders

```python
retail_df['invoice_date'] = retail_df['invoice_date'].dt.strftime('%Y%m')
# Cancelled orders report a negative value in the df we also filter for 0 values
filtered_cancels = retail_df[retail_df['total_amount'] < 0].drop_duplicates('invoice_no')
cancels = filtered_cancels.groupby('invoice_date')['invoice_no'].count()
# Total orders including the canceled are present 
orders = retail_df.drop_duplicates('invoice_no').groupby('invoice_date')['invoice_no'].count()

monthly_orders = pd.concat([orders - 2 * cancels, cancels], axis=1) 
monthly_orders.columns = ['Placement', 'Cancellation']
```

```python
monthly_orders.plot.bar(figsize=(15,10))
plt.xlabel('Month')
```

## Monthly Sales

```python
monthly_sales = (retail_df[retail_df['invoice_no'].str.isdigit()]
                 .groupby(by='invoice_date')['total_amount'] ## aggregation to get by month
                 .sum())
monthly_sales.plot.line()    
plt.xlabel('Month')
```

## Monthly Sales Growth

```python
monthly_growth = monthly_sales.pct_change() * 100 ## shows change month by month

monthly_growth.plot.line()
plt.xlabel('Month')
plt.ylabel('Monthly Growth (%)')
plt.title('Monthly Sales Growth')
plt.show()
```

## Monthly Active Users

```python
monthly_active = retail_df.groupby('invoice_date')['customer_id'].nunique()
monthly_active.plot.bar(figsize=(10, 7))
plt.xlabel('Month')
plt.show()
```

## New and Existing Users



```python
## Changing to a format to groupby 
retail_df['invoice_date'] = pd.to_datetime(retail_df['invoice_date'], format='%Y%m')

## identifying unique users by the id month by month
retail_df['first_purchase_month'] = retail_df.groupby('customer_id')['invoice_date'].transform('min').dt.to_period('M')
new_users = retail_df[retail_df['first_purchase_month'] == retail_df['invoice_date'].dt.to_period('M')]
new_users_count = new_users.groupby('invoice_date')['customer_id'].nunique()

# identifying existing users after first purchase month
previous_users = retail_df[retail_df['first_purchase_month'] < retail_df['invoice_date'].dt.to_period('M')]
previous_users_count = previous_users.groupby('invoice_date')['customer_id'].nunique()
```

```python
merged_data = pd.DataFrame({'New Users': new_users_count, 
            'Previous Users': previous_users_count}).fillna(0)
merged_data.plot.bar(figsize=(15, 10))

plt.xlabel('Month')
plt.legend(labels=['New Users', 'Existing Users'])
```

## Finding RFM

```python
reference_date = retail_df['invoice_date'].max()  # this is the date to reference for recency factor

rfm = retail_df.groupby('customer_id').agg({
    'invoice_date': lambda x: (reference_date - x.max()).days,  # check for recency value
    'invoice_no': 'count',                                      # check for frequency value
    'total_amount': 'sum'                                       # check for monetary value
}).reset_index()

rfm.rename(columns={'invoice_date': 'Recency', 'invoice_no': 'Frequency', 'total_amount': 'Monetary'}
           ,inplace=True)
rfm
```

## RFM Segmentation

```python
# RFM scores using quantiles 
rfm['recency_score'] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1])
rfm['frequency_score'] = pd.qcut(rfm['Frequency'].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
rfm['monetary_score'] = pd.qcut(rfm['Monetary'], 5, labels=[1, 2, 3, 4, 5])

# Create segment mapping with reference to RFM segmentation - business use
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Lose',
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}

rfm['segment'] = (rfm['recency_score'].astype(str) + rfm['frequency_score'].astype(str)).replace(seg_map, regex=True)

result = rfm.groupby(by='segment').agg({
    'Recency': ['mean', 'count'],  # Aggregating function similar to sql
    'Frequency': ['mean'],
    'Monetary': ['mean']
})

result
```
